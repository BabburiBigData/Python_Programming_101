```python
#Write a program to create a new string made of the middle three characters of an input string.
def get_3_middle_chars(str1):
    print(f"The string is {str1}")
    m=len(str1)//2
    new_string=str1[m-1:m+2]
    print(f"The middle 3 characters are:{new_string}")
get_3_middle_chars("JhonDipPeta")
```

```python
def find_digits_chars_symbols(sample_str):
    char_count = 0
    digit_count = 0
    symbol_count = 0
    for char in sample_str:
        if char.isalpha():
            char_count += 1
        elif char.isdigit():
            digit_count += 1
        # if it is not letter or digit then it is special symbol
        else:
            symbol_count += 1

    print("Chars =", char_count, "Digits =", digit_count, "Symbol =", symbol_count)

sample_str = "P@yn2at&#i5ve"
print("total counts of chars, Digits, and symbols \n")
find_digits_chars_symbols(sample_str)
```

```python
s1 = "Abc"
s2 = "Xyz"
s1_length=len(s1)
s2_length=len(s2)
result=''
s2=s2[::-1]
length=s1_length if s1_length>s2_length else s2_length
for i in range(length):
    if i<s1_length:
        result+=s1[i]
    if i<s2_length:
        result+=s2[i]
print(result)
```

```python
def string_balance_test(s1, s2):
    flag = True
    for char in s1:
        if char in s2:
            continue
        else:
            flag = False
    return flag


s1 = "Yn"
s2 = "PYnative"
flag = string_balance_test(s1, s2)
print("s1 and s2 are balanced:", flag)

s1 = "Ynf"
s2 = "PYnative"
flag = string_balance_test(s1, s2)
print("s1 and s2 are balanced:", flag)
```

```python
def sum_avg(str1):
    s=0
    c=0
    for i in str1:
        if i.isdigit():
            s+=int(i)
            c+=1
    avg=s/c
    print(avg,s)
    
sum_avg("PYnative29@#8496")
```

```python
def remove_char(str1):
    n=''
    for i in str1:
        if i.isdigit():
            n+=i
        else:
            continue
    n=int(n)
    print(n)

str1 = 'I am 25 years and 10 months old'
print(f"The string is {str1}")            
remove_char(str1)   
```

```python
str1 = "Emma25 is Data scientist50 and AI Expert"
print(f"The string is {str1}")  
temp=str1.split(" ")
print(temp)
res=[]
for item in temp:
    if any(char.isdigit() for char in item) and any(char.isalpha() for char in item):
        res.append(item)
    else:
        continue
print(res)
```
